````python
#Defining global functions for input validation

import sys

def valid_inputs(count, data_type, positive_only=False):
    while True:
        
        type_name = "number"
        
        if(data_type == int):
            type_name = "integer"
        elif(data_type == float):
            type_name = "float"
        elif(data_type == str):
            type_name = "string"
            
        user_input = input(f"Enter {count} input{'s' if count > 1 else ''} of type {type_name} separated by a space: ")
    
        if(user_input.strip() == ""):
            sys.exit("No input provided. Exiting.")
            
        try:
            inputs = user_input.split() #no .split(" ") needed, default splits on whitespace
            
            if(len(inputs) != count): 
                raise ValueError(f"You entered {len(inputs)} input{'s' if count > 1 else ''}. Please enter exactly {count} input{'s' if count > 1 else ''}.")
            
            result = list(map(data_type, inputs))   
            
            if(positive_only and any(num <= 0 for num in result)):
                raise ValueError(f"Entered input{'s' if count > 1 else ''} must be positive.")     
            
            return result;
        
        except ValueError as e:
            print(f"Invalid input. {e}")
            
````

## Program 28: Second Largest (Distinct)

## Program 30: List Palindrome Check

**Problem**: Read `N` and a list of `N` integers. Print `Palindrome` if the list reads the same forward and backward, else print `Not Palindrome`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- `Palindrome` or `Not Palindrome`

**Sample Input**:
```
5
1 2 3 2 1
```
**Sample Output**:
```
Palindrome
```


````python
print("PROBLEM 30: LIST PALINDROME CHECK".center(50, "-"))

print("Please enter an integer.")
number = valid_inputs(1, int)[0]
print(f"You entered: {number}")
print(f"Please enter {number} integer(s).")
numbers = valid_inputs(number, int)
print(f"You entered: {numbers}")

#print(f"{'PALINDROME' if numbers == numbers[::-1] else 'NOT PALINDROME'}") #shorter way using slicing

len_numbers = len(numbers)
for i in range(0, len_numbers//2, 1):
    if(numbers[i] != numbers[len_numbers - 1 - i]):
        print("Not palindrome.")
        break   
else:
    print("Palindrome.")
````
